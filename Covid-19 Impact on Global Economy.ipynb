### Number of country

```python
df2["COUNTRY"].value_counts()
```

Most countries occuring 294 times in data set

#### We are going to create a dataframe which contains single occurance of each country

```python
code = df2["CODE"].unique().tolist()
country = df2["COUNTRY"].unique().tolist()
population = df2["POP"].unique().tolist()
gdp = df2["GDPCAP"].unique().tolist()
hdi = []
tc = []
td = []
sti = []

for i in country:
    hdi.append(((df2.loc[df2["COUNTRY"] == i, "HDI"]).sum())/294)
    tc.append((df1.loc[df1["location"] == i, "total_cases"]).sum())
    td.append((df1.loc[df1["location"] == i, "total_deaths"]).sum())
    sti.append((df2.loc[df2["COUNTRY"] == i, "STI"]).sum()/294)
```

```python
data = pd.DataFrame(list(zip(code,country,population,tc,td,sti,gdp,hdi)),
                           columns = ["Country_code", "Country", "Total_Population", 
                                      "Total_Cases","Total_deaths","Stringency_Index",
                                      "GDP", "Human_Dev_Index"]
                       )
data.head()
```

```python
data["Country"].value_counts()
```

```python
data.isna().sum()
```

#### Sort dataframe by Total cases in descending order

```python
top_10_case=data.sort_values(by='Total_Cases', ascending=False)[:10]
top_10_case
```

**Maximuum Cases reported in Brazil Country.**

### Top 10 Country by Highest Covid-19 Cases

```python
fig = px.bar(top_10_case, x = 'Country', y = 'Total_Cases',
                 hover_data = ['Total_deaths','GDP'],
                 labels = {'Total_deaths':'Total deaths'}, height = 400
             )
fig.show()
```

```python
fig = px.scatter_geo(top_10_case, color="Country",locations = "Country_code",
                     hover_name="Country", size="Total_Cases",
                     projection="natural earth")
fig.show()
```

### Total Covid Cases %

```python
fig = px.pie(top_10_case, values='Total_Cases', names='Country', title='Total Covid Cases of the World') 
fig.show()
```

### Total Case & Total Deaths

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=top_10_case["Country"],
    y=top_10_case["Total_Cases"],
    name='Total_Cases',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=top_10_case["Country"],
    y=top_10_case["Total_deaths"],
    name='Total_deaths',
    marker_color='lightsalmon'
))
fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
```

### Total Case & Deaths Percentage

```python
cases = data["Total_Cases"].sum()
deceased = data["Total_deaths"].sum()

labels = ["Total Cases", "Total Deaths"]
values = [cases, deceased]

fig = px.pie(data, values=values, names=labels, 
             title='Percentage of Total Cases and Deaths', hole=0.5)
fig.show()
```
