### Exploratory Data Analysis

#### Change the Data type of "ActivityDate" column

```python
df["ActivityDate"] = pd.to_datetime(df["ActivityDate"],format="%m/%d/%Y")
df.dtypes
```

### How many minutes a day a person is active?

```python
df['TotalMinutes']=df['VeryActiveMinutes']+df['FairlyActiveMinutes']+df['LightlyActiveMinutes']
df.head()
```

### Relationship between Calories burnt & Total Steps

```python
figure=px.scatter(data_frame=df,
                 x='Calories', y='TotalSteps',
                 trendline='ols',
                 title='Relationship between Calories burnt & Total Steps')
figure.show()
```

**Trend is Positive that means as Total Steps increases then more Calories will burn.**

### Acvtivey Status by minutes a day

```python
label = ["Very Active Minutes", "Fairly Active Minutes", 
         "Lightly Active Minutes", "Inactive Minutes"]
counts = df[["VeryActiveMinutes", "FairlyActiveMinutes", 
               "LightlyActiveMinutes", "SedentaryMinutes"]].mean()
colors = ['gold','lightgreen', "pink", "blue"]

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Total Active Minutes')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```

- 81.3% minutes of the day are Inactive.
- 15.8% minutes of the day are lightly active.
- 1.11% minutes of the day are failry active.
- 1.74% minutes of the day are very active.

### Activity Status by Day

```python
df["Day"] = df["ActivityDate"].dt.day_name()
print(df["Day"].head())
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=df["Day"],
    y=df["VeryActiveMinutes"],
    name='Very Active',
    marker_color='black'
))
fig.add_trace(go.Bar(
    x=df["Day"],
    y=df["FairlyActiveMinutes"],
    name='Fairly Active',
    marker_color='red'
))
fig.add_trace(go.Bar(
    x=df["Day"],
    y=df["LightlyActiveMinutes"],
    name='Lightly Active',
    marker_color='blue'
))
fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
```

### Inactive Status by day

```python
day = df["Day"].value_counts()
label = day.index
counts = df["SedentaryMinutes"]
colors = ['gold','lightgreen', "pink", "blue", "skyblue", "cyan", "orange"]

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Inactive Minutes Daily')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```

- Thursday & then Monday are the most Inactive days.

### Calories burnt on each Day

```python
calories = df["Day"].value_counts()
label = calories.index
counts = df["Calories"]
colors = ['gold','lightgreen', "pink", "blue", "skyblue", "cyan", "orange"]

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Calories Burned Daily')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```

- More Calories burnt on Tuesday & then on Monday.
