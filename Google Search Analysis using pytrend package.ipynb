### Import required library

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
from pytrends.request import TrendReq
trends = TrendReq()
```

### Load the Data from Pytrends Library

```python
trends.build_payload(kw_list=['Python'])
data=trends.interest_by_region()
data=data.sort_values(by='Python', ascending=False)
data=data.head(10)
data
```

**In country China the keyword "Python" searched most of the times.**

### Plot the Searches on Bar plot

```python
data1=data.reset_index()
fig=px.bar(data_frame=data1,
          x='geoName',
          y='Python',
          title='Python Keyword Searched in Google')
fig.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Python'])
data = data.interest_over_time()
data
```

```python
data1=data.reset_index()
fig=px.line(data_frame=data1,
           x='date',
           y='Python',
           title='Total Google searches for keyword Python')
fig.show()
```

### Conclusion:
**Trend of the Search "python" Increasing over a period of time.**
